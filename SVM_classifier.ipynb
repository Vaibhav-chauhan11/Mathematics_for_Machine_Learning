```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
dir(iris)
```

```python
iris.feature_names
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
df['target'] = iris.target
df.head()
```

```python
iris.target_names
```

```python
df[df.target==2].head()
```

```python
df['flower_name']= df.target.apply(lambda x: iris.target_names[x])
df.head()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
df0 = df[df.target==0]
df1 = df[df.target==1]
df2= df[df.target==2]
df.head(2)
```

```python
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'],color='red',marker='*')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],color='blue',marker='.')
```

```python
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'],color='red',marker='*')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],color='blue',marker='.')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['target','flower_name'], axis = 'columns')
X.head()
```

```python
y = df.target
X_train, X_test, y_train , y_test = train_test_split(X,y,test_size= 0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.svm import SVC
model = SVC()  #(*, C=1.0, kernel='rbf', degree=3, gamma='scale', coef0=0.0, shrinking=True, probability=False, tol=0.001, cache_size=200, class_weight=None, verbose=False, max_iter=-1, decision_function_shape='ovr', break_ties=False, random_state=None)

model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
