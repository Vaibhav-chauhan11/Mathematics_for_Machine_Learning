```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(6)
x = 3 * np.random.rand(100,1)
y = 4 + 2*x + np.random.randn(100,1)
```

```python
plt.plot(x,y,'bo')
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(0,3)
plt.ylim(0,15)
plt.show()
```

```python
x_p = np.c_[np.ones((100,1)),x]
```

```python
x_p.shape
```

```python
x_p
```

```python
theta_best = np.linalg.inv(x_p.T.dot(x_p)).dot(x_p.T).dot(y)
```

```python
theta_best
```

```python
x_test = np.array([[0],[3]])
x_test_p = np.c_[np.ones((2,1)),x_test]
y_pred = x_test_p.dot(theta_best)
y_pred
```

```python
plt.plot(x_test,y_pred,'r-',label='Regression line')
plt.plot(x,y,'bo')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 1. Create Sample Data
# You can replace this with your own data
X = np.array([1, 2, 3, 4, 5, 6, 7, 8]).reshape(-1, 1)
y = np.array([2, 3.5, 3, 5.5, 6, 7, 6.5, 8])

model = LinearRegression()
model.fit(X, y)

y_model_pred = model.predict(X)

```

```python
#Define Own Line

my_slope = 0.85
my_intercept = 1.5
# Calculate the y-values for your custom line
y_my_line = my_slope * X + my_intercept
plt.figure(figsize=(10, 6))

# Ploting the original data points
plt.scatter(X, y, color='blue', label='Actual Data Points')

# Ploting the model's regression line
plt.plot(X, y_model_pred, color='red', linewidth=2, label=f"Model's Line (Best Fit)")

# Ploting  custom line
plt.plot(X, y_my_line, color='green', linestyle='--', linewidth=2, label=f"Your Line")

plt.title('Linear Regression: Model vs. Your Line')
plt.xlabel('X value')
plt.ylabel('y value')
plt.legend()
plt.grid(True)
plt.show()

print(f"Model's Slope (Coefficient): {model.coef_[0]:.2f}")
print(f"Model's Intercept: {model.intercept_:.2f}")
print("---")
print(f"Your Slope: {my_slope}")
print(f"Your Intercept: {my_intercept}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=load_diabetes()
df
```

```python
dataset=pd.DataFrame(df.data)
```

```python
dataset
```

```python
dataset.columns=df.feature_names
```

```python
dataset.head()
```

```python
#Independent Feature and dependent feature
```

```python
x=dataset
y=df.target
```

```python
y
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
X_train
```

```python
## staandardizing the dataset
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
## cross validation
from sklearn.model_selection import cross_val_score
```

```python
regression= LinearRegression()
regression.fit(X_train,y_train)
```

```python
mse=cross_val_score(regression,X_train,y_train,scoring='neg_mean_squared_error',cv=10)
```

```python
np.mean(mse)
```

```python
## prediction
reg_pred = regression.predict(X_test)
```

```python
reg_pred
```

```python
import seaborn as sns
sns.displot(reg_pred-y_test,kind='kde')
```

```python
from sklearn.metrics import r2_score
```

```python
score=r2_score(reg_pred,y_test)
```
